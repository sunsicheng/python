## 数组的创建

```python

#array:将输入的数据转换为ndarray，会进行copy
#asarray：将输入的数据转换为ndarray，如果输入本身是ndarray，则不会进行copy


import  numpy as np

data=[1,2,3,4,5,6]
print(id(data))

arr1=np.array(data)
print(id(arr1))

arr2=np.asarray(arr1)
print(id(arr2))


arr3=np.arange(-10,20,2)
print(arr3)
```

  ## 创建随机数数组
random.rand()：返回给定形状的数组，用 [0, 1) 上均匀分布的随机样本填充。
random.randint()：返回给定形状的数组，用从低位(包含)到高位(不包含)上均匀分布的随机整数填充。
random.uniform()：返回给定形状的数组，用从低位(包含)到高位(不包含)上均匀分布的随机浮点数填充。
random.randn()：返回给定形状的数组，用标准正态分布(均值为0，标准差为1)的随机样本填充。

```python
arr1=np.random.rand(2,3)
arr2=np.random.randint(0,10,(2,3))
arr3=np.random.randn(2,3)
print(arr1)
print(arr2)
print(arr3)
```

## Numpy的基本函数

```python
## 创建数组
import numpy as np

arr=np.arange(-10,10,2)
print(arr)
print(np.abs(arr))
print(np.ceil(18.8))
print(np.floor(18.8))
print(np.isnan(np.array(['1','2',None,'5'],dtype=float)))  #### NumPy 为了支持科学计算，会在需要时自动把 None 当成 NaN 处理，但只在 float 类型中可行,int中将报错
print(np.where(arr.size>2,'Y','N'))  ## 三元运算符
```
